```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data = pd.read_csv('Allergen_Status_Of_Food_Products.csv')
data['Prediction'].fillna('Unknown', inplace=True)
data['Sweetener'].fillna('Unknown', inplace=True) 
data['Fat/Oil'].fillna('Unknown', inplace=True)      
data['Seasoning'].fillna('Unknown', inplace=True)    
data['Allergens'].fillna('Unknown', inplace=True)   
```

```python
import pandas as pd
import category_encoders as ce
```

```python
# Ensure your dataset has the expected categorical columns
cat_cols = ['Food Product', 'Main Ingredient', 'Sweetener', 'Fat/Oil', 'Seasoning', 'Allergens']

# Define a target column (this is required for supervised encoding)
# Ensure your dataset has this column; replace 'Target' with the actual target column name
target_col = 'Prediction'

if target_col not in data.columns:
    raise ValueError(f"Target column '{target_col}' not found in the dataset!")

# Initialize the LeaveOneOutEncoder
encoder = ce.LeaveOneOutEncoder(cols=cat_cols)

# Fit and transform the dataset (target variable is required for this encoder)
data_encoded = encoder.fit_transform(data[cat_cols], data[target_col])

# Merge the encoded data with the original dataset
final_data = pd.concat([data.drop(columns=cat_cols), data_encoded], axis=1)

# Display the encoded dataset
print(final_data.head())
```
